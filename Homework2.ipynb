```python
import pandas as pd
```

```python
file='purchase_data.csv'

df_hero = pd.read_csv(file)
df_hero = df_hero.rename(columns={"Purchase ID":"Purchase_ID",
                        "SN":"Player",
                        "Item ID":"Item_ID",
                        "Item Name":"Item_Name"})
df_hero.head()
```

```python
#Player Count.
    #Total Number of Players
Players=df_hero['Player'].value_counts()
Players.count()
```

```python
#Gender Demographics
    #Percentage and Count of Male Players
    #Percentage and Count of Female Players
    #Percentage and Count of Other / Non-Disclosed
#Creamos un data frame que contenga a los Players unicos
df_gender = df_hero.groupby(['Gender']).nunique()
df_gender = df_gender['Player']
df_gender.to_frame()
#Reiniciamos index
df_gender = df_gender.reset_index(level=0)
df_gender
#Obtenemos el porcentaje de los Players unicos
df_percentage = df_hero['Gender'].value_counts(df_hero['Player'].nunique())
df_percentage.to_frame()
df_percentage = df_percentage.reset_index(level=0)
df_percentage= df_percentage.rename(columns={"index":"Gender",
                                            "Gender":"Player"})
#Utilizamos merge para juntar ambos data frame
gender_df=df_gender.merge(df_percentage,on='Gender')

gender_df= gender_df.rename(columns={"Player_x":"Player",
                                            "Player_y":"%Percentage"})
gender_df
```

```python
#Purchasing Analysis (Gender)

#The below each broken by gender
        #Purchase Count
        #Average Purchase Price
        #Total Purchase Value
        #Average Purchase Total per Person by Gender
#Calculamos promedio, contamos y sumamos informacion de los generos
df_groupCPurchase = df_hero[['Gender','Purchase_ID']].groupby(['Gender']).count()
df_groupMPurchase = df_hero[['Gender','Price']].groupby(['Gender']).mean()
df_groupSPurchase = df_hero[['Gender','Price']].groupby(['Gender']).sum()
df_groupSPerson = df_hero[['Gender','Player']].groupby(['Gender']).nunique()
#Obtenemos el promedio por persona
avgT=df_groupSPurchase['Price']/df_groupSPerson['Player']
#Agregamos nuevas columnas con los datos que nos piden
df_groupCPurchase['Avg Price']= df_groupMPurchase['Price']
df_groupCPurchase['Total Revenue']= df_groupSPurchase['Price']
df_groupCPurchase['Avg Total Purchase per Person'] = avgT
df_groupCPurchase = df_groupCPurchase.reset_index(level=0)
df_groupCPurchase = df_groupCPurchase.rename(columns={"Purchase_ID":"Purchase Count"})
df_groupCPurchase
```

```python
#Crando bins para data frame
df_hero['bin'] = pd.cut(df_hero['Age'], bins=[0,9,14,19,24,29,34,39,50],labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39','40+'])
```

```python
#Contando jugadores por bin
df_bin = df_hero[['bin','Player']].groupby(['bin']).nunique()
df_bin =df_bin['Player']
df_bin=df_bin.to_frame()
df_bin = df_bin.reset_index(level=0)
df_bin= df_bin.rename(columns={"Count of Players":"Bin",
                                "Player":"Count of Players"})

df_bin['Percentage of Players'] = (df_bin['Count of Players']/576)*100

df_bin
```

```python
#Obtenemos promedio,total y numero de players
df_binPurchase = df_hero[['bin','Purchase_ID']].groupby(['bin']).nunique()
df_binPurchaseprice = df_hero[['bin','Price']].groupby(['bin']).mean()
df_binPurchasetotalprice = df_hero[['bin','Price']].groupby(['bin']).sum()
df_bincount = df_hero[['bin','Player']].groupby(['bin']).nunique()
avgTbin=df_binPurchasetotalprice['Price']/df_bincount['Player']



#Agregamos nuevas columnas con los datos que nos piden
df_binPurchase['Avg Price']= df_binPurchaseprice['Price']
df_binPurchase['Total Revenue']= df_binPurchasetotalprice['Price']
df_binPurchase['Avg Total Purchase per Person'] = avgTbin
df_binPurchase = df_binPurchase.rename(columns={"Purchase_ID":"Purchase Count"})
df_binPurchase=df_binPurchase[['Purchase Count','Avg Price','Total Revenue','Avg Total Purchase per Person']]
df_binPurchase
```

```python

df_PlayerCount = df_hero[['Player','Purchase_ID']].groupby(['Player']).count()
df_PlayeMean = df_hero[['Player','Price']].groupby(['Player']).mean()
df_Playertotal = df_hero[['Player','Price']].groupby(['Player']).sum()

df_PlayerCount['Average Purchase Price']=df_PlayeMean
df_PlayerCount['Total Purchase Value']=df_Playertotal
df_PlayerCount=df_PlayerCount.sort_values(by=['Total Purchase Value'],ascending=False)
df_PlayerCount.head()
```

```python
df_itemCount = df_hero[['Item_ID','Item_Name','Purchase_ID']].groupby(['Item_Name']).count()
df_itemMin = df_hero[['Item_ID','Item_Name','Price']].groupby(['Item_Name']).min()
df_itemtotal = df_hero[['Item_ID','Item_Name','Price']].groupby(['Item_Name']).sum()

df_itemtotal
```

```python
df_itemCount['Item Price']=df_itemMin['Price']
df_itemCount['Total Purchase Value']=df_itemtotal['Price']
df_itemCount=df_itemCount.sort_values(by=['Purchase_ID'],ascending=False)
df_itemCount=df_itemCount[['Purchase_ID','Item Price','Total Purchase Value']]
df_itemCount.head()
```
